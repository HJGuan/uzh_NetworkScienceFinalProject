```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("dom_citypairs_web.csv")
```

```python
df
```

```python
df['Total Distance_(km)'] = df['Aircraft_Trips'] * df['Distance_GC_(km)']
```

```python
drop_list = ['Month', 'Passenger_Load_Factor', 'RPKs', 'ASKs', 'Month_num','Distance_GC_(km)']
```

```python
df.drop(drop_list,inplace = True, axis = 1)
```

```python
df
```

```python
df2 = df.groupby(['City1','City2','Year'], as_index =False).sum()
```

```python
df2['City Distance_(km)'] = df2['Total Distance_(km)']/df2['Aircraft_Trips']
```

```python
df2
```

```python
df2 = df2.dropna(axis = 0)
```

```python
df2
```

```python
df2.to_csv("new.csv")
```

```python
path = 'yearly_data/'
```

```python
year = df2['Year'].unique()
```

```python
for y in year:
    filtered_df = df2[df2['Year'] == y]
    filtered_df.to_csv(path + str(y) + ".csv")
```
